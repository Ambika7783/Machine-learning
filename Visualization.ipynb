Graphs:Regex--->regular expression
Regex is like a mini-language for finding patterns in text.

Why use Regex?

*Validate text (e.g., check if an email or phone number is valid)
*Search text (e.g., find all words starting with a capital letter)
*Replace text (e.g., change all dates to a standard format)
*extract text (e.g., pick out numbers from a paragraph)

Common Regex Patterns
Pattern	Matches
.	Any single character (except newline)
\d	Any digit (0–9)
\D	Any non-digit
\w	Any word character (letters, digits, underscore)
\W	Any non-word character
\s	Any whitespace (space, tab, newline)
\S	Any non-whitespace
^	Start of string
$	End of string
*	0 or more repeats
+	1 or more repeats
?	0 or 1 repeat (optional)
{n}	Exactly n times
{n,m}	Between n and m times
[...]	Any character in brackets

Real-life Uses

*Form validation (emails, phone numbers, passwords)

*Data cleaning (remove unwanted characters, extract info)

*Log analysis (find error codes, IP addresses)

*Web scraping (extract data from HTML/text)

```python
import re

text = "My phone number is 9876543210 and my email is test123@mail.com"

# Find all numbers
print(re.findall(r"\d+", text))
# ['9876543210', '123']

# Find all words
print(re.findall(r"\w+", text))
# ['My', 'phone', 'number', 'is', '9876543210', 'and', 'my', 'email', 'is', 'test123', 'mail', 'com']

# Check if email is valid
pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b"
print(bool(re.match(pattern, "test123@mail.com")))
# True
```

```python
re.findall(emailPattern,sentence)  #What is re.findall()?

It’s a function from Python’s re (regular expression) module.

It returns all non-overlapping matches of a pattern in a string.

The result is always a list of strings.

re.findall(pattern, string, flags=0)
pattern → the regex pattern you want to search for

string → the text you want to search in

flags → optional (like re.IGNORECASE, re.MULTILINE)
re.search() → returns first match (or None)

re.match() → only checks at the beginning of string

re.findall() → returns all matches in a list

re.finditer() → returns an iterator of match objects (with positions)
```

```python
'''1. Line Graph → trends over time
used when your x-axis is time (or ordered data).'''
import matplotlib.pyplot as plt

years = [2018, 2019, 2020, 2021, 2022]
sales = [100, 120, 90, 150, 180]

plt.plot(years, sales, marker="o")
plt.title("Sales Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()

```

```python
'''2. Bar Graph → compare categories
Used to compare groups.'''
brands = ["Ford", "BMW", "Toyota", "Honda"]
sales = [200, 150, 300, 180]

plt.bar(brands, sales, color="skyblue")
plt.title("Car Sales by Brand")
plt.xlabel("Brand")
plt.ylabel("Sales")
plt.show()
```

```python
'''3. Histogram → distribution of values

Used to check how values are spread (e.g., test scores).'''
import numpy as np

scores = np.random.randint(40, 100, 200)  # 200 random test scores
plt.hist(scores, bins=10, color="orange", edgecolor="black")
plt.title("Distribution of Test Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

```python
''' 4. Scatter Plot → relationship between variables

Used when comparing two continuous variables.'''
weight = [150, 160, 170, 180, 190, 200]
mpg = [35, 33, 30, 28, 25, 22]

plt.scatter(weight, mpg, color="red")
plt.title("Car Weight vs MPG")
plt.xlabel("Weight (kg)")
plt.ylabel("Miles per Gallon")
plt.show()



```

```python
''' 5. Pie Chart → proportions

Used to show percentages or parts of a whole.'''

sizes = [40, 25, 20, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Fruit Market Share")
plt.show()
```

```python
'''build a Box Plot graph step by step.
1. Simple Box Plot (single variable) '''
import matplotlib.pyplot as plt

# Example data: exam marks
marks = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100, 45, 50, 30]

plt.boxplot(marks)
plt.title("Box Plot of Student Marks")
plt.ylabel("Marks")
plt.show()

#👉 This shows median, quartiles, whiskers, and outliers of student marks.
```

Matplotlib is a low-level, comprehensive data visualization library in Python.
It provides control over every aspect of a figure: from plots, axes, ticks, labels, markers, line styles, colors, to full customization of charts.

It’s the foundation library on which higher-level libraries like Seaborn and Pandas plotting are built.
Key Features

Supports many types of plots: line, bar, scatter, histogram, pie, etc.

Very flexible & customizable (you can tweak every element of the plot).

Works well with NumPy arrays and Pandas DataFrames.

Can save plots in multiple formats (PNG, JPG, PDF, SVG).

Useful for scientific visualization and publication-quality figures.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Line plot
plt.plot(x, y, marker='o', color='blue', linestyle='dashed', label='y=2x')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot (Matplotlib)")
plt.legend()
plt.grid(True)
plt.show()
```

Violin plot:Boxplot+frequency

A Violin Plot is a type of data visualization that shows the distribution of numerical data across different categories.
It is similar to a boxplot, but it also includes a kernel density estimation (KDE), making it look like a “violin”.

Key Features

Combines boxplot (median, quartiles, outliers) with distribution shape (density).

Useful for comparing distributions between multiple groups.

Wider sections of the violin = more data points in that range.

Narrower sections = fewer data points.

x="day" → categories on x-axis (Thur, Fri, Sat, Sun).

y="total_bill" → numeric variable (distribution shown as violin).

hue="sex" → splits violin by gender.

split=True → both groups shown side-by-side in the same violin.


```python
sns.violinplot(data=df,y='tip')
plt.show()
```

PAIR PLOT:
A Pair Plot (also called a scatterplot matrix) is a grid of plots that shows the relationship between multiple numerical variables in a dataset.

It plots scatter plots for each pair of variables.

Shows histograms (or KDEs) on the diagonal to display individual variable distributions.

Useful for exploring datasets quickly, checking correlations, and spotting patterns.

Why Pair Plots are Useful

Great for exploratory data analysis (EDA).

Lets you visualize all variable relationships at once.

Helps detect clusters, correlations, and outliers.

👉 In short: Pair Plot = many scatter plots + histograms in one grid.




```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
iris = sns.load_dataset("iris")

# Create pair plot
sns.pairplot(iris, hue="species", diag_kind="kde")

plt.show()
```

iris dataset has 4 numerical features (sepal length, sepal width, petal length, petal width) + category (species).

hue="species" → points are colored by species (Setosa, Versicolor, Virginica).

diag_kind="kde" → diagonal plots show smooth density instead of histograms.

HEATMAP:A Heatmap is a data visualization technique that shows values in a matrix format using colors.

Each cell’s color represents the magnitude of the value.

Useful for visualizing correlation matrices, confusion matrices, or large tables

Heatmaps are Useful

Quickly identify strong positive/negative correlations.

Great for large datasets where tables are hard to read.

Often used in EDA, machine learning, and statistical analysis.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")

# Select only numeric columns
iris_numeric = iris.select_dtypes(include='number')

# Compute correlation matrix
corr = iris_numeric.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Only)")
plt.show()

```

iris.corr() → computes correlation between numerical features.

annot=True → writes the correlation values inside each cell.

cmap="coolwarm" → color gradient (blue = low, red = high).

linewidths=0.5 → draws lines between cells for better readability.
